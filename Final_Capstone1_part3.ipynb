```python
import pandas as pd
import numpy as np
import random
import requests
from bs4 import BeautifulSoup
from IPython.display import Image 
from IPython.core.display import HTML 
from IPython.display import display_html
```

```python
source = requests.get('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=900271985').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
```

```python
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
dfs = pd.read_html(tab)
df=dfs[0]
```

```python
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['Postcode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2
```

```python
# Shape of data frame
df2.shape
```

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
df_temp = df2.set_index('Postcode')
coors_temp = lat_lon.set_index('Postal Code')
df_coors = pd.concat([df_temp, coors_temp], axis=1, join='inner')

df_coors.index.name = 'PostalCode'
df_coors.reset_index(inplace=True)

print(df_coors.shape)
df_coors.head()
```

<h1>Looking for text = "Toronto"</h1>

```python
df3 = df_coors[df_coors['Borough'].str.contains('Toronto',regex=False)]
df3
```

<h1>Using Folium to show in start zoom the "East Toronto"</h1>

```python
import folium
map_toronto = folium.Map(location=[43.662744,-79.321558],zoom_start=13)

for lat,lng,borough,neighbourhood in zip(df3['Latitude'],df3['Longitude'],df3['Borough'],df3['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='green',
    fill_opacity=0.4,
    parse_html=False).add_to(map_toronto)
map_toronto
```
